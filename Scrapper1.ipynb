```python
import requests, bs4
```

```python
request1=requests.get('https://www.ebay.com/itm/156172001386?epid=0&itmmeta=01HWNG5C7YCD74634CCT8WMPJM&hash=item245c93a06a:g:GzoAAOSwcY5mIDIH')
request1
```

```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
```

```python
#reviews
review1=soup.find('div',{'class':'fdbk-container__details__comment'})
print(review1.get_text())
```

```python
reviews=soup.findAll('div',{'class':'fdbk-container__details__comment'})
for review in reviews:
  print(review.get_text());
```

```python
#ratings
rating1=soup.findAll('div',{'class':'vim x-rating-details'})
for rate in rating1:
  print(rate.get_text() + "/n/n")
```

```python
#customer_name
customer_name=soup.findAll('div',{'class':'x-review-section__author'})
for cust in customer_name:
  print(cust.get_text())
```

```python
review2=soup.findAll('span',{'class':'ux-textspans'})
for review in review2:
  print(review.get_text());
```

```python
review3=soup.findAll('span',{'class':'INLINE_LINK'})
for review in review3:
  print(review.get_text());
```

```python
review4=soup.findAll('div',{'class':'fdbk-container__details__comment'})
for review in review4:
  print(review.get_text() + "\n\n");
```

```python
review5=soup.find('h2',{'class':'recs-title-label  '})
print(review5.get_text());
```
